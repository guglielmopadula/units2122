```python
import sklearn
import numpy as np
from sklearn.datasets import load_digits
import pyro
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
import torch
import pyro.distributions as dist
import pyro.optim as optim
from pyro.infer import SVI, Trace_ELBO
import pandas as pd
import seaborn as sns
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import numpy as np
from pyro.infer import Predictive
import torch.distributions.constraints as constraints
dataset = load_digits()
x, y = dataset.data, dataset.target
X=torch.tensor((x-x.min())/(x.max()-x.min()))
y=torch.tensor(y)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
sns.set_style("darkgrid")
figsize=(10,4)
pyro.set_rng_seed(0)

pyro.clear_param_store()
def model(x, y):
    k=len(torch.unique(y))
    n_observations, n_predictors = x.shape
    w = pyro.sample("w", dist.Normal(torch.zeros(k,n_predictors), torch.ones(k,n_predictors)))
    b = pyro.sample("b", dist.Normal(torch.zeros(k),torch.ones(k)))
    yhat = torch.sigmoid(torch.matmul(x.float(),w.T) + b)
    with pyro.plate("data", n_observations):
        y = pyro.sample("y", dist.Categorical(yhat), obs=y)
        

def guide(x, y=torch.zeros(10)):
    k=len(torch.unique(y))
    n_observations, n_predictors = x.shape
    w_loc = pyro.param("w_loc", torch.rand(k,n_predictors))
    w_scale = pyro.param("w_scale", torch.rand(k,n_predictors),
    constraint=constraints.positive)
    w = pyro.sample("w", dist.Normal(w_loc, w_scale))
    b_loc = pyro.param("b_loc", torch.rand(k))
    b_scale = pyro.param("b_scale", torch.rand(k),constraint=constraints.positive)
    b = pyro.sample("b", dist.Normal(b_loc, b_scale))
    
log_reg_svi = SVI(model=model, guide=guide,
optim=optim.Adam({'lr' : 0.0002}),loss=Trace_ELBO())
losses = []
for step in range(10000):
    loss = log_reg_svi.step(X_train, y_train.float())/len(X_train)
    losses.append(loss)
fig, ax = plt.subplots(figsize=figsize)
ax.plot(losses)
ax.set_title("ELBO loss");
```

```python
w = pyro.get_param_store()["w_loc"]
b = pyro.get_param_store()["b_loc"]
def predict_class(x):
    out = torch.sigmoid(torch.matmul(x.float(),w.T) + b)
    return torch.argmax(out,dim=1)
```

```python
pred=predict_class(X_test)
```

```python
print("Total accuracy is", accuracy_score(y_test, pred))
matrix = confusion_matrix(y_test, pred)
acc=matrix.diagonal()/matrix.sum(axis=1)
for i in range(10):
    print("accuracy of class",i,"is",acc[i])
```
