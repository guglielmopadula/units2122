## Text
Implement the empirical cumulative distribution function $F_{X}(x)=\operatorname{cdf}($ dist $, \mathrm{x}, \mathrm{n}$ samples) taking as inputs a pyro.distributions object dist, corresponding to the distribution of $X$, a real value $x$ and the number of samples n_samples.

The function cdf(dist, $x$, n_samples) should return the value of $F_{X}$ at $x$ and also plot the cdf.

Suppose that $X \sim \operatorname{Exp}(0.5)$. Using your function, plot $F_{X}$ and compute $F_{X}(x=2)$.



```python
import torch
import math
import pyro
pyro.set_rng_seed(1)
import matplotlib.pyplot as plt

def cdf(dist, x, n_samples):
    vec=dist.sample(torch.Size([n_samples]))
    vec,_=torch.sort(vec)
    ind=torch.DoubleTensor([i for i in range(n_samples)])/n_samples
    plt.plot(vec.tolist(),ind.tolist())
    v=torch.sum(vec<=x,dtype=torch.double)/n_samples
    return v.item()

v=cdf(pyro.distributions.Exponential(0.5),2,1000000)
print("Fx is", v)
```

Suppose the heights of female students are normally distributed with unknown mean $\mu$ and known variance $6^{2}$.

2. Suppose $\mu \sim N(\mu_{0},\sigma_{0}^{2})$\\
$$ f(\mu \mid \underline{x}) \propto f(\mu) f(\underline{x} \mid \mu)=f(\mu) f\left(x_{1} \mid \mu\right) f\left(x_{2} \mid \mu\right) \ldots f\left(x_{n} \mid \mu\right)
=\frac{1}{\sqrt{2 \pi \sigma_{0}^{2}}} \exp \left\{\frac{-\left(\mu-\mu_{0}\right)^{2}}{2 \sigma_{0}^{2}}\right\} \times \prod_{i=1}^{n} \frac{1}{\sqrt{2 \pi \sigma^{2}}} \exp \left\{\frac{-\left(x_{i}-\mu\right)^{2}}{2 \sigma^{2}}\right\}=\frac{1}{(2 \pi)^{\frac{(n+1)}{2}} \sqrt{\sigma_{0}^{2} \sigma^{2 n}}} \exp \left\{\frac{-\mu^{2}+2 \mu \mu_{0}-\mu_{0}^{2}}{2 \sigma_{0}^{2}}\right. \left.-\sum_{i=1}^{n} \frac{x_{i}^{2}-2 \mu x_{i}+\mu^{2}}{2 \sigma^{2}}\right\}$$ $$\propto \exp \left\{\frac{-\mu^{2}\left(\sigma^{2}+n \sigma_{0}^{2}\right)+2 \mu\left(\mu_{0} \sigma^{2}+\sigma_{0}^{2} x_{1}+\cdots+\sigma_{0}^{2} x_{n}\right)-\left(\mu_{0}^{2} \sigma^{2}+\sigma_{0}^{2} x_{1}^{2}+\cdots+\sigma_{0}^{2} x_{n}^{2}\right)}{2 \sigma_{0}^{2} \sigma^{2}}\right\}$$ $$\propto \exp \left\{-\frac{\left(\mu-\frac{\mu_{0} \sigma^{2}+\sum_{i=1}^{n} \sigma_{0}^{2} x_{i}}{\sigma^{2}+n \sigma_{0}^{2}}\right)^{2}}{2 \frac{\sigma_{0}^{2} \sigma^{2}}{\sigma^{2}+n \sigma_{0}^{2}}}\right\}$$
so $\mu \mid \underline{x} \sim N\left(\frac{\mu_{0} \sigma^{2}+\sum_{i=1}^{n} \sigma_{0}^{2} x_{i}}{\sigma^{2}+n \sigma_{0}^{2}}, \frac{\sigma^{2} \sigma_{0}^{2}}{\sigma^{2}+n \sigma_{0}^{2}}\right)$

```python
x=[174,158,194,167]
mu0=160
sigma0=3
sigma=6
s=sum(x)
n=4
mupos=(mu0*sigma**2+sigma0**2*s)/(sigma**2+n*sigma0**2)
sigmapos=math.sqrt((sigma**2*sigma0**2)/(sigma**2+n*sigma0**2))
t=torch.linspace(140,180,1000000)
distprior=pyro.distributions.Normal(mu0,sigma0)
distpos=pyro.distributions.Normal(mupos,sigmapos)
yprior=torch.exp(distprior.log_prob(t))
ypos=torch.exp(distpos.log_prob(t))
plt.plot(t.tolist(),yprior.tolist(),label="prior")
plt.plot(t.tolist(),ypos.tolist(),label="posterior")
plt.legend()
plt.show()
```
