# Exercise 1

```python
import torch
import math
import pyro
import seaborn as sns
pyro.set_rng_seed(1)
import pyro.distributions as dist
import matplotlib.pyplot as plt
from pyro.infer.mcmc import MCMC, HMC, NUTS
import warnings
warnings.filterwarnings("ignore")


def model(theta1_prior,theta2_prior, x, obs):
    theta1=pyro.sample('theta1',theta1_prior)
    theta2=pyro.sample('theta2',theta2_prior)
    yhat=(theta1+x**2)/(theta2*x)
    y=pyro.sample('y',dist.Normal(yhat,1),obs=obs)
    return y

x=torch.tensor([47,87,20,16,38,5])
y=torch.tensor([58.76,108.75,25.03,20.03,47.51,6.37])
```

```python
#Beta Beta


theta1_prior=dist.Beta(4,5)
theta2_prior=dist.Beta(3,2)
nuts_kernel = NUTS(model)
mcmc = MCMC(nuts_kernel, num_samples=500, warmup_steps=1000, num_chains=1)
posterior = mcmc.run(theta1_prior,theta2_prior,x,y)
mcmc_samples = mcmc.get_samples(group_by_chain=True)
theta1 = mcmc_samples["theta1"]
theta2 = mcmc_samples["theta2"]
n_chains, n_samples = theta1.shape
mcmc.summary()
sns.distplot(mcmc_samples['theta1'])
plt.xlabel("Theta1")
plt.show()
sns.distplot(mcmc_samples['theta2'])
plt.xlabel("Theta2")
plt.show()
print("Theta1 (Uniform)  converged Theta2 (Beta)  converged")

```

```python
#Unif Unif


x=torch.tensor([47,87,20,16,38,5])
y=torch.tensor([58.76,108.75,25.03,20.03,47.51,6.37])
theta1_prior=dist.Uniform(0.,1.)
theta2_prior=dist.Uniform(0.,1.)
hmc_kernel = HMC(model)
mcmc = MCMC(hmc_kernel, num_samples=500, warmup_steps=1000, num_chains=1)
posterior = mcmc.run(theta1_prior,theta2_prior,x,y)
mcmc_samples = mcmc.get_samples(group_by_chain=True)
theta1 = mcmc_samples["theta1"]
theta2 = mcmc_samples["theta2"]
n_chains, n_samples = theta1.shape
mcmc.summary()
sns.distplot(mcmc_samples['theta1'])
plt.xlabel("Theta1")
plt.show()
sns.distplot(mcmc_samples['theta2'])
plt.xlabel("Theta2")
plt.show()
print("Theta1 (Uniform) not converged Theta2 (Beta) not converged")
```

```python
theta1_prior=dist.Uniform(0,1)
theta2_prior=dist.Beta(2,2)
hmc_kernel = HMC(model)
mcmc = MCMC(hmc_kernel, num_samples=500, warmup_steps=1000, num_chains=1)
posterior = mcmc.run(theta1_prior,theta2_prior,x,y)
mcmc_samples = mcmc.get_samples(group_by_chain=True)
theta1 = mcmc_samples["theta1"]
n_chains, n_samples = theta1.shape
mcmc.summary()
sns.distplot(mcmc_samples['theta1'])
plt.xlabel("Theta1")
plt.show()
sns.distplot(mcmc_samples['theta2'])
plt.xlabel("Theta2")
plt.show()
print("Theta1 (Uniform) not converged Theta2  (Beta) not converged")
```

Not all the mcmc converged. The posterior distributions of $\theta_{2}$ are similar to a beta, even if the prior is uniform, while the distribution of $\theta_{1}$ are more similar to the respective prior.



# Exercise 2

From the theory of multivariable normal distribution we know that $X_{1}|X_{2}=x_{2}\sim N(\rho x_{2},1-\rho^{2}$) and $X_{2}|X_{1}=x_{1}\sim N(\rho x_{1},1-\rho^{2})$

```python
def gibbs_sampler(rho, warmup, iters):
    x1=torch.zeros(warmup+iters)
    x2=torch.zeros(warmup+iters)
    x1[0]=pyro.sample("start",dist.Uniform(0,1))
    for i in range(1,warmup+iters):
        x2[i]=pyro.sample("x2_{}".format(i),dist.Normal(rho*x1[i-1].item(),torch.sqrt(1-rho**2)))
        x1[i]=pyro.sample("x1_{}".format(i),dist.Normal(rho*x2[i].item(),torch.sqrt(1-rho**2)))
        
    return x1[warmup:],x2[warmup:]

theta1,theta2=gibbs_sampler(torch.tensor([0.5]),1000,500)
sns.distplot(theta1)
plt.xlabel("Theta1")
plt.show()
sns.distplot(theta2)
plt.xlabel("Theta2")
plt.show()
```
