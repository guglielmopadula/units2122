```python
import gurobipy as gp
import numpy as np
from gurobipy import GRB
```

```python
#Model has fixed size
def model(W):
    if W<=10:
        print("Error, by equation 15 we must have W>12")
        return -1
######################DATA PREPARATION######################
#Parameters taken by the paper
    FOC=1000
    MOB=10000
    OB=0.02
    OP=0.5
    OV=0.1
    AP=15000
    LP=2
    LR=0
    R=2
    RP=0.25
    CL=18000
    rW=0.00123
    V=4
#Parameters taken by the data (r1)
    N=16
    S=np.zeros(16, dtype=object)
    S[0]=[3,10,11]
    S[1]=[5,11]
    S[2]=[4,6,7]
    S[3]=[5,8]
    S[4]=[9,11]
    S[5]=[6,11]
    S[6]=[13,15]
    S[7]=[8,10]
    S[8]=[13,14,15]
    S[9]=[12,14]
    S[10]=[12,13]
    S[11]=[15]
    S[12]=[15]
    S[13]=[]
    S[14]=[]
    S[15]=[]
    M=np.zeros(16, dtype=object)
    for i in range(16):
        M[i]=[0,1,2]
    C=np.zeros([16,3],dtype=np.int64)
    C[0,0]=14
    C[0,1]=5
    C[0,2]=2
    C[1,0]=9
    C[1,1]=4
    C[1,2]=12
    C[2,0]=15
    C[2,1]=8
    C[2,2]=9
    C[3,0]=3
    C[3,1]=12
    C[3,2]=6
    C[4,0]=20
    C[4,1]=14
    C[4,2]=6
    C[5,0]=13
    C[5,1]=4
    C[5,2]=9
    C[6,0]=13
    C[6,1]=9
    C[6,2]=4
    C[7,0]=16
    C[7,1]=5
    C[7,2]=4
    C[8,0]=12
    C[8,1]=4
    C[8,2]=2
    C[9,0]=16
    C[9,1]=9
    C[9,2]=10
    C[10,0]=3
    C[10,1]=13
    C[10,2]=9
    C[11,0]=6
    C[11,1]=5
    C[11,2]=10
    C[12,0]=11
    C[12,1]=3
    C[12,2]=6
    C[13,0]=14
    C[13,1]=13
    C[13,2]=9
    C[14,0]=10
    C[14,1]=12
    C[14,2]=2
    C[15,0]=7
    C[15,1]=12
    C[15,2]=2
    T=np.zeros([16,3],dtype=np.int64)
    T[0,0]=1
    T[0,1]=1
    T[0,2]=2
    T[1,0]=3
    T[1,1]=7
    T[1,2]=10
    T[2,0]=2
    T[2,1]=9
    T[2,2]=10
    T[3,0]=3
    T[3,1]=8
    T[3,2]=9
    T[4,0]=3
    T[4,1]=6
    T[4,2]=8
    T[5,0]=3
    T[5,1]=7
    T[5,2]=10
    T[6,0]=1
    T[6,1]=2
    T[6,2]=2
    T[7,0]=3
    T[7,1]=5
    T[7,2]=9
    T[8,0]=1
    T[8,1]=2
    T[8,2]=10
    T[9,0]=1
    T[9,1]=1
    T[9,2]=5
    T[10,0]=5
    T[10,1]=7
    T[10,2]=10
    T[11,0]=1
    T[11,1]=3
    T[11,2]=6
    T[12,0]=4
    T[12,1]=4
    T[12,2]=4
    T[13,0]=1
    T[13,1]=1
    T[13,2]=9
    T[14,0]=2
    T[14,1]=6
    T[14,2]=7
    T[15,0]=3
    T[15,1]=9
    T[15,2]=10
      
    DC=np.zeros([N,3,W], dtype=np.int64)
    for i in range(W):
        DC[:,:,i]=C
#######Some calculations#########    
    DCsum=np.sum(C[:,0])#Equation 1 
    VOCsum=OV*DCsum# Equation 2
    FOCsum=DCsum*W# Equation 3
    PM=OP*(DCsum+VOCsum+FOCsum+MOB)#Equation 4
    BC=OB*(DCsum+VOCsum+FOCsum+MOB+PM)#Equation 5
    BP=DCsum+VOCsum+FOCsum+MOB+PM+BC#Equation 6
    MU=BP/DCsum#Equation 7
    DCk=np.zeros(W, dtype=np.int64)
    for k in range(1,W):
        for i in range(N):
            for j in M[i]:
                DCk[k]=DCk[k]+DC[i,j,k] #Equation 8
    VOCk=OV*DCk #Equation 9
    TEk=DCk+VOCk+FOC#Equation 10
    TEk[0]=MOB+BC#Equation 11 
    TEk.resize(int(np.ceil(W/R))*R+LP+LR)
    MMF=np.zeros([N,3])
    for i in range(N):
        for j in M[i]:
            MMF[i,j]=C[i,0]/C[i,j] #Equation 12
        
    EV=np.zeros([N,W])
    for i in range(N):
        for k in range(1,W):
            for j in M[i]:
                EV[i,k]=EV[i,k]+MMF[i,j]*DC[i,j,k] #equation 13
    
    I=np.zeros(int(np.ceil(W/R))*R+LP+1)
    for k in range(W):
        if k%R==0 & k<np.ceil(W/R):
            for i in range(N):
                for t in range(max(k-R+1,0),k+1):
                    I[k]=I[k]+MU*EV[i,t] #Equation 14
    P=np.zeros(int(np.ceil(W/R))*R+LP+1) 
    for k in range(R+LP,int(np.ceil(W/R))*R+LP+1):
        if (k-LP)%R==0 & k<=np.ceil(W/R)+LP:
            P[k]=(1-R)*I[k-LP]-AP/np.ceil(W/R)
    P[int(np.ceil(W/R))+R+LP+LR]=P[int(np.ceil(W/R))+R+LP+LR]+RP*BP #equation 15
    
    B=np.zeros(max(int(np.ceil(W/R))*R+LP+LR+1,int(np.ceil(W/V)*V))+1)
    IB=np.zeros(max(int(np.ceil(W/R))*R+LP+LR+1,int(np.ceil(W/V)*V))+1)
    B[0]=AP-TEk[0]
    for k in range(1,int(np.ceil(W/R))*R+LP):
        if k%V==0 or k==int(np.ceil(W/V)*V):
            for t in range(k-V+1,k):
                IB[k]=IB[k]+min(0,TEk[t]-(B[t-1]-P[t-1]))*rW
        B[k]=B[k-1]-TEk[k]+P[k-1]-IB[k]
    B[int(np.ceil(W/R))*R+LP+LR]=B[int(np.ceil(W/R))*R+LP]+RP*BP#equation 16-17 
    
    model=gp.Model()
    
    #Equation from 18 to 22 are not used
    x = model.addVars(N,N,W, vtype=GRB.BINARY, name="x")
    dc = model.addVars(N,W, name="dc")
    

    model.addConstrs(gp.quicksum(x[i,j,k] for j in M[i] for k in range(1,W))==1 for i in range(N))#equation 23
    model.addConstrs(gp.quicksum((k+T[i,j])*x[i,j,k] for j in M[i] for k in range(1,W))<gp.quicksum(k*x[q,j,k] for j in M[q] for k in range(1,W)) for q in S[i] for i in range(N)) #equation 24
    model.addConstrs(gp.quicksum((k+T[i,j])*x[i,j,k] for j in M[i] for k in range(1,W))<=W for i in range(N) if len(S[i])==0) #equation 25
    model.addConstrs(dc[i,k]==gp.quicksum(C[i,j]/T[i,j]*x[i,j,t] for t in range(max(1,k-T[i,j]),k) for j in M[i]) for i in range(N) for k in range(1,W)) #equation 26
    #model.addConstrs(TEk[k]==FOC+(1+OV)*gp.quicksum(dc[i,k] for i in range(N)) for k in range(W)) #equation 27
    model.addConstrs(EV[i,k]==gp.quicksum(MMF[i,j]*dc[i,k] for j in M[i]) for i in range(N) for k in range(1,W)) #equation 26: MODIFIED by me

    model.optimize()
    print(TEk[:W])
```

```python
model(20)
```
